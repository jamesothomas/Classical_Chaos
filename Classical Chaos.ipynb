Euler-Cromer Method

```python
'''
modified version of:
Simple Pendulum - EulerCromer Method
by Dr. Wood

James O. Thomas
'''
import numpy as np
import matplotlib.pylab as plt

# initialize variables
multiplier = 1.0
omega_init = 0
theta_init = 0.2
nsteps = int(200*multiplier)
g = 9.8
length = 1.0
dt = 0.1/multiplier

# new variables for the energy
m = 1      # kg  needed for the gravatational potential

omega = np.zeros(nsteps)
theta = np.zeros(nsteps)
energy = np.zeros(nsteps)
t = np.arange(0, nsteps*dt, dt)

omega[0] = omega_init
theta[0] = theta_init

for i in range(nsteps-1):
    omega[i+1] = omega[i] - (g/length)*theta[i]*dt
    theta[i+1] = theta[i] + omega[i+1]*dt           # Euler Cromer Method
    
energy = 0.5*m*(length*omega)**2+m*g*length*(1-np.cos(theta))

plt.figure(figsize = (8,6))
plt.plot(t,theta,'k-',label = r'$\theta $')
plt.plot(t,omega,'b-',label = r'$\omega $')
plt.plot(t, energy, 'r-', label = 'Total Energy')
plt.xlabel('Time (s)')
plt.ylabel(r'$\theta $ (radians) '+'\n or $\omega $ (radians/second) \n or energy (juoles)') 
# the r (Latex) thing doesnt work with \n (new line) so I just used two strings
plt.legend()
plt.show()
```

Comparison of Runge-Kutta and Euler-Cromer

```python
'''
modified version of:
Simple Pendulum - EulerCromer Method
by Dr. Wood

James O. Thomas
'''
import numpy as np
import matplotlib.pylab as plt

def RHS(y):
    # returns a vector of the derivatives
    return [y[1], -g/length*y[0]]

def Jrk4(f, h):
    
    k1 = RHS([f[0], f[1]])                               # use initial point
    k2 = RHS([f[0] + 0.5*k1[0]*h, f[1] + 0.5*k1[1]*h])   # one of the midpoints
    k3 = RHS([f[0] + 0.5*k2[0]*h, f[1] + 0.5*k2[1]*h])   # the other midpoint    
    k4 = RHS([f[0] + k3[0]*h, f[1] + k3[1]*h])           # the endpoint
    f_next = [f[0] + (k1[0] + 2*k2[0] + 2*k3[0] + k4[0])*h/6, f[1] + (k1[1] + 2*k2[1] + 2*k3[1] + k4[1])*h/6]
    return f_next

# initialize variables
multiplier = 1.0
omega_init = 0
theta_init = 0.2
nsteps = int(200*multiplier)
g = 9.8
length = 1.0
dt = 0.1/multiplier

# new variables for the energy
m = 1      # kg  needed for the gravatational potential

omega_euler = np.zeros(nsteps)
theta_euler = np.zeros(nsteps)
energy_euler = np.zeros(nsteps)
omega_euler1 = np.zeros(nsteps)
theta_euler1 = np.zeros(nsteps)
energy_euler1 = np.zeros(nsteps)
rk4 = np.zeros((nsteps,2))     # [theta, omega]
energy_rk4 = np.zeros(nsteps)
t = np.arange(0, nsteps*dt, dt)

omega_euler[0] = omega_init
theta_euler[0] = theta_init
omega_euler1[0] = omega_init
theta_euler1[0] = theta_init
rk4[0,:] = [theta_init, omega_init]

for i in range(nsteps-1):
    omega_euler[i+1] = omega_euler[i] - (g/length)*theta_euler[i]*dt
    theta_euler[i+1] = theta_euler[i] + omega_euler[i+1]*dt           # Euler Cromer Method
    omega_euler1[i+1] = omega_euler1[i] - (g/length)*theta_euler1[i]*dt
    theta_euler1[i+1] = theta_euler1[i] + omega_euler1[i]*dt           # Euler Method
    rk4[i+1] = Jrk4(rk4[i], dt)
    
# analytic solutions
# sines and cosines are switched from normal pendulum equation because of initial condition
theta_analytic = theta_init*np.cos(np.sqrt(g/length)*t)
omega_analytic = -theta_init*np.sqrt(g/length)*np.sin(np.sqrt(g/length)*t)

# calculate the energy
energy_euler = 0.5*m*(length*omega_euler)**2 + m*g*length*(1-np.cos(theta_euler))
energy_euler1 = 0.5*m*(length*omega_euler1)**2 + m*g*length*(1-np.cos(theta_euler1))
energy_rk4 = 0.5*m*(length*rk4[:,1])**2 + m*g*length*(1-np.cos(rk4[:,0]))   
energy_analytic = 0.5*m*(length*omega_analytic)**2 + m*g*length*(1-np.cos(theta_analytic))

plt.figure(figsize = (15,5))

ax1 = plt.subplot(131)
ax1.plot(t, omega_euler, label = 'Euler-Cromer')
ax1.plot(t, omega_euler1, label = 'Euler')
ax1.plot(t, rk4[:,1], label = 'rk4')
ax1.plot(t, omega_analytic, label = 'analytic')
ax1.set_ylabel('omega [rads/s]')
ax1.set_xlabel('time [s]')
ax1.set_ylim(-2,2)
l1 = plt.legend()
plt.title('omega')

ax2 = plt.subplot(132)
ax2.plot(t, theta_euler, label = 'Euler-Cromer')
ax2.plot(t, theta_euler1, label = 'Euler')
ax2.plot(t, rk4[:,0], label = 'rk4')
ax2.plot(t, theta_analytic, label = 'analytic')
ax2.set_ylabel('theta [radians]')
ax2.set_xlabel('time [s]')
ax2.set_ylim(-1,1)
l2 = plt.legend()
plt.title('theta')

ax3 = plt.subplot(133)
ax3.plot(t, energy_euler, label = 'Euler-Cromer')
ax3.plot(t, energy_euler1, label = 'Euler')
ax3.plot(t, energy_rk4, label = 'rk4')
ax3.plot(t, energy_analytic, label = 'analytic')
ax3.set_ylim(-2,2)
ax3.set_ylabel('energy [joules]')
ax3.set_xlabel('time [s]')
ax3.set_ylim(0.1,0.4)
l3 = plt.legend()
plt.title('energy')

plt.suptitle('Comparison of Euler and rk4 methods', size = 16)
plt.show()
```

Liapunov exponent

```python
'''
James O. Thomas
October 2015
'''

import matplotlib.pylab as plt
import numpy as np
from scipy.integrate import ode

def f1(t, y):
    return[y[1], -2*beta*y[1] -omega_0**2*np.sin(y[0]) + A1*np.sin(Omega*t)]      # [d(theta)/dt, d(omega)/dt]

def f2(t, y):
    return[y[1], -2*beta*y[1] -omega_0**2*np.sin(y[0]) + A2*np.sin(Omega*t)]       # for second pundulum

# define some parameters (similar to classical mechanics book)
Omega = 2*np.pi  # driving frequency
beta = 3/4*np.pi   # b/(2m)
omega_0 = 3*np.pi # natural frequency = sqrt(g/l)  
gamma1 = 1.18 # related to amplitude of driving force = F_0/(mg)
gamma2 = 1.16
A1 = gamma1*omega_0**2 # related to amplitude of driving force = gamma omega_0^2 = F_0/(ml)
A2 = gamma2*omega_0**2

# initial values
omega_init1 = 0.0
theta_init1 = 0.0
omega_init2 = 0.0
theta_init2 = 0.0
t_init = 0

multiplier = 100.
nsteps = int(20*multiplier)
dt = 2/multiplier

# initialize arrays and set initial values
t1 = np.zeros(nsteps)
t2 = np.zeros(nsteps)
omega1 = np.zeros(nsteps)
theta1 = np.zeros(nsteps)
omega2 = np.zeros(nsteps)
theta2 = np.zeros(nsteps)
t1[0] = t_init
omega1[0] = omega_init1
theta1[0] = theta_init1
t2[0] = t_init
omega2[0] = omega_init2
theta2[0] = theta_init2

# set up the integrators
r1 = ode(f1).set_integrator('dopri5')
r1.set_initial_value([theta_init1, omega_init1], t_init)
r2 = ode(f2).set_integrator('dopri5')
r2.set_initial_value([theta_init2, omega_init2], t_init)

# integrate the 1st pendulum
k = 1
while r1.successful() and k < nsteps:
    r1.integrate(r1.t + dt)
    t1[k] = r1.t                      # should be the same values as the second integrator
    theta1[k] = r1.y[0]
    omega1[k] = r1.y[1]
    k = k + 1
    
# integrate the 2nd pendulum
k = 1
while r2.successful() and k < nsteps:
    r2.integrate(r2.t + dt)
    t2[k] = r2.t                      # t values should be the same, but just in case
    theta2[k] = r2.y[0]
    omega2[k] = r2.y[1]
    k = k + 1

# calculate the stuff for the Lyapunov exponent
lndtheta = np.log(np.abs(theta1 - theta2)+0.01)
m, b = np.polyfit(t1,lndtheta,1)
# print(m,b)

plt.figure(figsize=(10,5))
ax = plt.subplot(1,2,1)
ax.plot(t1, theta1, label = r'$\theta _1$')
ax.plot(t2, theta2, label = r'$\theta _2$')
ax.legend(loc = 'lower left')
ax.set_xlabel('time [s]')
ax.set_ylabel(r'$\theta $ [rads]')
ax.set_title('Chaotic Pendulums')

ax2 = plt.subplot(1,2,2)
ax2.plot(t1,lndtheta, label = r'$ln (\Delta \theta )$')
ax2.plot(t1, m*t1 + b, label = 'linear fit')
ax2.set_title('Lyapunov exponent')
ax2.legend(loc = 'lower left')
ax2.set_xlabel('time [s]')
ax2.set_ylabel(r'$ln (\Delta \theta )$')
ax2.text(20,-2, ' y=mt+b for linear fit \n m = '+str(m)+' \n b = '+str(b))
plt.show()
```

Poincare Maps

```python
'''
James O. Thomas
'''

import time
import matplotlib.pylab as plt
import numpy as np
from scipy.integrate import ode

def f(t, y):                           # y[theta, omega]
    return[y[1], -g/l*np.sin(y[0]) - q*y[1] + F_D*np.sin(omega_drive*t)]      # [d(theta)/dt, d(omega)/dt]

# define some parameters like Giordano book
g = 9.8
l = 9.8
omega_drive = 2./3.
dt = 0.005
q = 0.5
F_D = 1.2

# initial values
omega_init = 0.0
theta_init = 0.2
t_init = 0

nsteps = int(3e7)

omega = []   # empty list2
theta = []   

# set up the integrators
r = ode(f).set_integrator('dopri5')
r.set_initial_value([theta_init, omega_init], t_init)

# integrate the pendulum
start = time.time()
twopi = 2*np.pi
k = 0
n = 1
while r.successful() and k < nsteps:
    r.integrate(r.t + dt)
    t_val = r.t                      
    theta_val = r.y[0]
    omega_val = r.y[1]
    # only keep the points that are (almost) in phase with the driving force
    if np.abs(t_val - 2*n*np.pi/omega_drive) < dt/2:   # extra factor of 2 from Giordano book
        omega.append(omega_val)
        theta.append(theta_val%twopi)
        n = n + 1
    k = k + 1
    # show that it is doing something
    if k%3e5<0.01:
        print('.', end = '')
        
print('finished')
stop = time.time()
print('elapsed time is ', stop - start)
# change the domain from (0, 2pi) to (-pi, pi)
for i in range(len(theta)):
    if theta[i] > np.pi:
        theta[i] = theta[i] - 2*np.pi

print('number of points found = ',len(omega))
plt.figure(figsize = (10,5))
ax = plt.subplot(1,2,1)
ax.set_title('Chaotic Attractor')
ax.plot(theta,omega,'k,')
ax.set_xlabel(r'$\theta $ [radians]')
ax.set_xlim(-np.pi,np.pi)
ax.set_ylabel(r'$\omega $ [rad/s]')

ax1 = plt.subplot(1,2,2)
ax1.set_title('Zoomed in')
ax1.set_xlabel(r'$\theta $ [radians]')
ax1.set_ylabel(r'$\omega $ [rad/s]')
ax1.set_xlim(1.9,np.pi)
ax1.set_ylim(-1.5,0)
ax1.plot(theta,omega, 'k,')
plt.show()

## save the data to a text file so I dont have to regenerate the data to replot
#file = open("theta_v_omega.txt", "w")
#for index in range(len(theta)):
#    file.write(str(theta[index]) + " " + str(omega[index]) + "\n")
#file.close()
```

Henon-Hiles

```python
import numpy as np
import matplotlib.pyplot as plt

def Henon_Heiles(x,y):
    return 0.5*(x**2 + y**2) + x**2*y - 1./3*y**3
    
endpt = 2
stpsize = 0.01

x_val = np.arange(-endpt,endpt+stpsize,stpsize)
y_val = np.arange(-endpt,endpt+stpsize,stpsize)

#contours = [3,2,1,0.5,0.167,0.1,0.05]
contours = [0.05,0.1,0.167,0.5,1,2,3]
V = np.zeros([len(x_val),len(y_val)])

for x in range(len(x_val)):
    for y in range(len(y_val)):
        V[x,y] = Henon_Heiles(x_val[x],y_val[y])

V = np.transpose(V)	
X,Y = np.meshgrid(x_val,y_val)	
plt.figure(figsize = (10,10))
#plt.title('Equipotential Contours')
CS = plt.contour(X,Y,V,contours)
plt.clabel(CS,inline=1, fontsize=10)
plt.xlabel('x')
plt.ylabel('y')
plt.text(-0.1,-1.8,'Ridge')
plt.text(-0.1,1.8,'Valley')
plt.show()
```

Feigenbaum Plot

```python
import numpy as np
import matplotlib.pylab as plt

def logistic_map(r,x):
    x = r*x*(1-x)
    return x
dstep = 0.005
r = np.arange(1,4+dstep,dstep) # the r values to use
x = np.empty(len(r))        # empty array for x values
x.fill(0.5)                  # fill the x with the initial value
                    
steps = 1000                  # the number of times to run the logistic map

fig = plt.figure(figsize = (10,5))
ax = fig.add_subplot(1,2,1)
ax1 = fig.add_subplot(1,2,2)

for i in range(steps):
    x = logistic_map(r,x)        # let it settle down
for i in range(steps):
    x = logistic_map(r,x)        # plot the behavior
    ax.plot(r,x,"k,")
    ax1.plot(r,x,"k,")

ax.set_title('The Feigenbaum Plot')
ax.set_xlabel('the r value')
ax1.set_title('Zoomed In')
ax1.set_xlabel('the r value')
ax1.set_xlim(2.9,3.6)
ax1.set_ylim(0.65,0.9)
plt.show()
```

Mandelbrot Set

```python
import numpy as np
import matplotlib.pylab as plt
    
def Mandelbrot(c):
    z = 0 + 0j
    for i in range(100):
        z = z**2 + c
        if abs(z) > 2:
            break
    if abs(z) >= 2:
        return False           # this c value does not lie in the madelbrot set
    else:
        return True           # this c value does lie in the mandelbrot set

steps = 500
step_size = 4./steps
counter = 0 # just for the print('working...') statement
plt.figure()

mand = np.empty((steps,steps))
real_axis = np.arange(-2,2+step_size,step_size)
imag_axis = np.arange(-2,2+step_size,step_size)

for r in range(steps):
    c_real = real_axis[r]
    for i in range(steps):
        c_imag = imag_axis[i]
        c = complex(c_real,c_imag)
        out = Mandelbrot(c)
        mand[i,r] = out
        
plt.figure()
plt.title('Colored Mandelbrot Set')
plt.xlabel('Real axis')
plt.ylabel('Imaginary axis')
plt.imshow(mand, extent = [-2,2,-2,2])
plt.show()
```

```python
import numpy as np
import matplotlib.pylab as plt

def Mandelbrot(c, n=100):
    z = 0 + 0j
    for i in range(n):
        z = z**2 + c
        if abs(z) > 2:
            n = i        # return the number of steps taken to get out of the Mandelbrot set
            break
    return n                # if in the Mandelbrot set, return 100

n_steps = 500
step_size = 4./n_steps

# first initialize the array for storing the values
mand = np.empty((n_steps,n_steps))
real_axis = np.arange(-2,2+step_size,step_size)
imag_axis = np.arange(-2,2+step_size,step_size)

for r in range(n_steps):
    c_real = real_axis[r]
    for i in range(n_steps):
        c_imag = imag_axis[i]
        c = complex(c_real,c_imag)
        out = Mandelbrot(c)
        if out > 0:                # special care since log(0) is undefined
            out = np.log10(out)
        mand[i,r] = out

        
plt.figure()
plt.title('Colored Mandelbrot Set')
plt.xlabel('Real axis')
plt.ylabel('Imaginary axis')
plt.imshow(mand, extent = [-2,2,-2,2])
plt.show()
```
